```python
def open_kamus_prepro(x):
	kamus={}
	with open(x,'r') as file :
		for line in file :
			slang=line.replace("'","").split(':')
			kamus[slang[0].strip()]=slang[1].rstrip('\n').lstrip()
	return kamus

kamus_slang=open_kamus_prepro('C:/Users/sinag/SKRIPSIKU/slangwords.txt')
```

```python
def slangword(text):  
    sentence_list = text.split()
    new_sentence = []
    for word in sentence_list:
      for candidate_replacement in kamus_slang:
        if candidate_replacement == word:
          word = word.replace(candidate_replacement, kamus_slang[candidate_replacement])
      new_sentence.append(word)
    return " ".join(new_sentence)

data['text'] = data['text'].apply(lambda x: slangword(x))
data.head()
```

```python
# Proses Tokenizing (memecah kalimat menjadi kata)
def tokenizing(text):
    text = nltk.tokenize.word_tokenize (text)
    return text
data['text'] = data ['text'].apply(tokenizing)
data.head()
```

```python
# Menambah daftar stopword 
stopword = open('C:/Users/sinag/SKRIPSIKU/stopwords.txt',"r",encoding="utf-8", errors='replace')
stopword = [t.strip('\n').strip() for t in stopword]

# Proses Filtering/Stopword (membuang kata yang tidak dibutuhkan)
def stopword_removal(text):
    filtering = stopword
    x = []
    data = []
    def myFunc(x):
        if x in filtering:
            return False
        else:
            return True
    fit = filter(myFunc, text)
    for x in fit:
        data.append(x)
    return data
data['text'] = data ['text'].apply(stopword_removal)
data.head()
```

```python
# proses stemming
! pip install Sastrawi
from sklearn.pipeline import Pipeline
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean=[]
    d_clean=" ".join(do)
    print(d_clean)
    return d_clean
data['text'] = data['text'].apply(stemming)
data.head()
```

```python
def open_kamus_prepro(x):
	kamus={}
	with open(x,'r') as file :
		for line in file :
			slang=line.replace("'","").split(':')
			kamus[slang[0].strip()]=slang[1].rstrip('\n').lstrip()
	return kamus

kamus_slang=open_kamus_prepro('C:/Users/sinag/SKRIPSIKU/slang_pemerintah.txt')
```

```python
def slangwords(text):  
    sentence_list = text.split()
    new_sentence = []
    for word in sentence_list:
      for candidate_replacement in kamus_slang:
        if candidate_replacement == word:
          word = word.replace(candidate_replacement, kamus_slang[candidate_replacement])
      new_sentence.append(word)
    return " ".join(new_sentence)

data['text'] = data['text'].apply(lambda x: slangword(x))
data.head()
```

```python
data_clean = data_clean.astype({'label' : 'category'})
data_clean = data_clean.astype({'text' : 'string'})
data_clean.dtypes
```

```python
def preprocess_data(text):
    text = casefolding(text)
    text = cleansing(text)
    text = slangword(text)
    text = slangwords(text)
    text = tokenizing(text)
    text = stopword_removal(text)
    text = stemming(text)
    
    return text
```

```python
# method untuk cleaning dokumen berupa array
def clean_with_loop(arr):
    hasil = []
    progress = tqdm(arr)
    for item in progress:
        progress.set_description("Membersihkan dokumen")
        cleaned = preprocess_data(item)
        cleaned = ' '.join(cleaned)
        hasil.append(cleaned)
    return hasil
```
