```python
def open_kamus_prepro(x):
	kamus={}
	with open(x,'r') as file :
		for line in file :
			slang=line.replace("'","").split(':')
			kamus[slang[0].strip()]=slang[1].rstrip('\n').lstrip()
	return kamus

kamus_slang=open_kamus_prepro('C:/Users/sinag/belajar/slangwords.txt')
```

```python
def slangword(text):  
    sentence_list = text.split()
    new_sentence = []
    for word in sentence_list:
      for candidate_replacement in kamus_slang:
        if candidate_replacement == word:
          word = word.replace(candidate_replacement, kamus_slang[candidate_replacement])
      new_sentence.append(word)
    return " ".join(new_sentence)

data['text'] = data['text'].apply(lambda x: slangword(x))
data.head(10)
```

```python
# Proses Tokenizing (memecah kalimat menjadi kata)
def token(text):
    text = nltk.tokenize.word_tokenize (text)
    return text
data['text'] = data ['text'].apply(token)
data.head()
```

```python
# Menambah daftar stopword 
stopword = open('C:/Users/sinag/Documents/coba/login_flask/stopwords.txt',"r",encoding="utf-8", errors='replace')
stopword = [t.strip('\n').strip() for t in stopword]

# Proses Filtering/Stopword (membuang kata yang tidak dibutuhkan)
def stopword_removal(text):
    filtering = stopword
    x = []
    data = []
    def myFunc(x):
        if x in filtering:
            return False
        else:
            return True
    fit = filter(myFunc, text)
    for x in fit:
        data.append(x)
    return data
data['text'] = data ['text'].apply(stopword_removal)
data.head(10)
```

```python
# proses stemming
! pip install Sastrawi
from sklearn.pipeline import Pipeline
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean=[]
    d_clean=" ".join(do)
    print(d_clean)
    return d_clean
data['text'] = data['text'].apply(stemming)
data.head(20)
```

```python
data_clean = data_clean.astype({'label' : 'category'})
data_clean = data_clean.astype({'text' : 'string'})
data_clean.dtypes
```

```python
# PROSES TF IDF
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data_clean['text'].astype('U'))

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data_clean['text'].astype('U'))
print(text_tf)
```

```python
# Import cosine_similarity metrics
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# compute similarity using cosine similarity
cos_sim=cosine_similarity(text_tf, text_tf)
print(cos_sim)
```

```python
# splitting data latih 70 : uji 30
import collections, numpy
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cos_sim, data_clean['label'], test_size=0.3, random_state=33)
print("Jumlah Data Uji:", X_test.shape)
print("Jumlah Data Latih:",X_train.shape)

pos = (y_test == 1).sum()
neg = (y_test == 0).sum()
postrain = (y_train == 1).sum()
negtrain = (y_train == 0).sum()
total = pos + neg
print("Jumlah data uji dengan sentimen positif:", pos)
print("Jumlah data uji dengan sentimen negatif:",neg)
print("Jumlah data latih dengan sentimen positif:", postrain)
print("Jumlah data latih dengan sentimen negatif:",negtrain)
data_clean['label'].value_counts()
```

```python
# perform algoritma KNN
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=9).fit(X_train, y_train)
predicted = clf.predict(X_test)
print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('===============================================\n')
tn, fp, fn, tp = confusion_matrix(y_test, predicted).ravel()
print("TN:", tn)
print("FP:", fp)
print("FN:", fn)
print("TP:", tp)
print(classification_report(y_test, predicted, zero_division=0))
print('===============================================\n')
print("Hasil Klasifikasi Sentimen Analisis Shopee:")

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
predknn = knn.predict(X_test)
print("Confusion Matrix for K Neighbors Classifier:")
print(confusion_matrix(y_test,predknn))
print("Score: ",round(accuracy_score(y_test,predknn)*100,2))
print("Classification Report:")
print(classification_report(y_test,predknn)) 
```
